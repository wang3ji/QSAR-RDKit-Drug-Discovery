# QSAR
Quantitative Structure–Activity Relationship


QSAR models try to predict the biological activity or property of a molecule based on its chemical structure
1. Input: molecular structure (usually converted into descriptors or fingerprints)
2. Output: predicted activity/property (e.g., dolubility, toxicity,enzyme inhibition, drug efficacy)

🔹 Descriptors

Numerical values that describe molecular properties:

1D descriptors: Molecular weight, number of atoms, number of rings.

2D descriptors: LogP (hydrophobicity), hydrogen bond donors/acceptors, topological indices.

3D descriptors: Molecular surface area, 3D conformations.

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
```

```python
mol = Chem.MolFromSmiles("CCO")
mw = Descriptors.MolWt(mol)
logp = Descriptors.MolLogP(mol)
print(f"MW={mw}, LogP = {logp}")
```

🔹 Molecular Fingerprints

Binary vectors representing presence/absence of substructures or patterns.

Example: Morgan fingerprint (circular fingerprints) in RDKit.

Used to measure similarity between molecules or as input for ML models.

```python
from rdkit.Chem import AllChem
fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=1024)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Example descriptors (molecular weight only for simplicity)
X = np.array([[Descriptors.MolWt(Chem.MolFromSmiles(s))] for s in ['CCO','CC','CCC','CCN']])
y = np.array([0.8, 0.5, 1.0, 0.9])  # some activity values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)
print(pred)
```
